```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
template_style = "plotly_dark"  # You can choose from 'plotly', 'ggplot2', 'seaborn', 'simple_white', etc.

layout_config = dict(
    template=template_style,
    font=dict(family="Arial", size=14, color="white"),
    title_x=0.5,
    margin=dict(l=40, r=40, t=60, b=40)
)
```

```python
df = pd.read_csv('../data/fake_job_postings.csv')

df.head()
```

```python
df.info()
```

```python

print(df.isnull().sum())
before = df.isnull().sum()

# Percentage missing
print((df.isnull().sum() / len(df)) * 100)
```

```python
missing = df.isnull().sum().reset_index()
missing.columns = ["Column", "Missing Values"]

fig = px.bar(
    missing,
    x="Column",
    y="Missing Values",
    title="Missing Values by Column"
)
fig.update_layout(**layout_config)
fig.show()
```

```python
missing_pct = (
    df.isnull().mean()*100
).reset_index()

missing_pct.columns = ["Column", "Percentage"]

fig = px.bar(
    missing_pct,
    x="Column",
    y="Percentage",
    title="Percentage of Missing Values",
)
fig.update_layout(**layout_config)
fig.show()
```

```python
df.drop(columns=["department", "salary_range"], inplace=True)

df.dropna(subset=["description"], inplace=True)


```

```python
text_columns = [
    "company_profile",
    "requirements",
    "benefits"
]

for col in text_columns:
    df[col] = df[col].fillna("Not Provided")
```

```python
cat_columns = [
    "location",
    "employment_type",
    "required_experience",
    "required_education",
    "industry",
    "function"
]

for col in cat_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
after = df.isnull().sum()

report = pd.DataFrame({
    "Before": before,
    "After": after
})

print(report)
```

```python
industry_count = (
    df["industry"]
    .value_counts()
    .head(10)
    .reset_index()
)

industry_count.columns = ["Industry", "Count"]

fig = px.bar(
    industry_count,
    x="Industry",
    y="Count",
    title="Top 10 Industries"
)
fig.update_layout(**layout_config)
fig.show()
```

```python
employment = (
    df["employment_type"]
    .value_counts()
    .reset_index()
)

employment.columns = ["Type", "Count"]

fig = px.pie(
    employment,
    names="Type",
    values="Count",
    title="Employment Type Distribution"
)
fig.update_layout(**layout_config)
fig.show()
```

```python
fraud = (
    df["fraudulent"]
    .value_counts()
    .reset_index()
)

fraud.columns = ["Type", "Count"]

fraud["Type"] = fraud["Type"].map({
    0:"Legitimate",
    1:"Fraudulent"
})

fig = px.pie(
    fraud,
    names="Type",
    values="Count",
    title="Fraudulent vs Legitimate Jobs"
)
fig.update_layout(**layout_config)
fig.show()
```

```python
remote = (
    df["telecommuting"]
    .value_counts()
    .reset_index()
)

remote.columns = ["Remote", "Count"]

remote["Remote"] = remote["Remote"].map({
    0:"On-site",
    1:"Remote"
})

fig = px.pie(
    remote,
    names="Remote",
    values="Count",
    title="Remote vs On-site Jobs"
)

fig.update_layout(**layout_config)
fig.show()
```

```python
locations = (
    df["location"]
    .value_counts()
    .head(15)
    .reset_index()
)

locations.columns = ["Location", "Count"]

fig = px.bar(
    locations,
    x="Location",
    y="Count",
    title="Top Job Locations"
)

fig.update_layout(**layout_config)
fig.show()
```
